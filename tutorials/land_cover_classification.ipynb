# PyEumap - Tutorial

```python
import sys
sys.path.append('../')

from pyeumap.overlay import SpaceOverlay, SpaceTimeOverlay
from pathlib import Path

import geopandas as gpd
import pandas as pd
import os
```

```python
#data_dir = '/home/leandro/Tmp/pilot_tiles/sweden_22497'
data_dir = '/home/leandro/Tmp/pilot_tiles/croatia_9529'

fn_points = os.path.join(data_dir,'training_samples.gpkg')
col_date = 'survey_date'

dir_layers = os.path.join(data_dir,'images')
dir_timeless_layers = os.path.join(dir_layers, 'timeless')

csv_output = os.path.join(data_dir, 'training_samples_overlayed.csv')
gpkg_output =  os.path.join(data_dir, 'training_samples_overlayed.gpkg')
```

# Space Overlay

```python
spcOverlay = SpaceOverlay(fn_points, dir_timeless_layers)
timelessData = spcOverlay.run()
```

# Space-Time Overlay

```python
sptOverlay = SpaceTimeOverlay(fn_points, col_date, dir_layers)
timeData = sptOverlay.run()
```

# Merging and save the data

```python
timelessColumns = list(set(timelessData.columns) - set(timeData.columns)) + ['overlay_id']
data = timeData.merge(timelessData[timelessColumns])
```

```python
data.to_csv(csv_output)
data.to_file(gpkg_output,  driver="GPKG")
```
